## 한 게임의 기록 저장하기

```python
# 우리가 저장할 데이터프레임의 컬럼 이름이 안 이쁨.
# 컬럼 이름 이쁘게 만들어주는 함수
def replace_numbers(e, number_position_mapping):
    for num, position in number_position_mapping.items():
        e = e.replace(num, position)
    return e
```

```python
game_summaries = []
fail_to_load_games = []

start_time = time.time()
for game_id in game_ids:
    time.sleep(1.5)
    print(game_id, end = " ")
    try:
        a_game_timeline = pd.DataFrame(API_hander.get_game_timeline_information(game_id)['info']['frames'])
        print(len(a_game_timeline))
    except:
        fail_to_load_games.append(game_id)
        continue

    # 한 클래스가 한 게임의 정보를 요약하기에, 게임이 바뀔 때 마다 객체 선언 및 초기화
    GameSummary = EventRecorder()

    a_game_summaries = []

    each_team_positions = GameSummary.find_positions(a_game_timeline)
    blue_number_position_mapping = {f"{number}" : f"blue_{each_team_positions[0][number - 1]}" for number in range(1,6)}
    red_number_position_mapping = {f"{number + 5}" : f"red_{each_team_positions[1][number - 1]}" for number in range(1,6)}
    number_position_mapping = red_number_position_mapping | blue_number_position_mapping
    # 매 판 각 팀의 포지션 번호를 기록

    for index in range(1, len(a_game_timeline)):
        a_minute_events = pd.DataFrame(a_game_timeline.loc[index, "events"])
        a_minute_events.apply(GameSummary.update_score_board, axis = 1) # CS를 제외한 모든 정보를 기록해줌.
        pd.Series(range(1,11)).apply(lambda x : GameSummary.update_score_board_from_CS(x, index, a_game_timeline))

        if index >= 5: # 5분 이상의 정보만 기록
            record_dicts = [GameSummary.record_KDACS, 
                            GameSummary.record_building, 
                            GameSummary.record_turret_plate, 
                            GameSummary.record_object,
                            GameSummary.record_level,
                            GameSummary.record_blue_win]
            a_game_summary_until_minute = pd.concat([pd.DataFrame([record_dict]) for record_dict in record_dicts], axis = 1)
            new_col_name = [replace_numbers(e, number_position_mapping) for e in list(a_game_summary_until_minute.columns)]
            a_game_summary_until_minute.columns = new_col_name
            
            a_game_summaries.append(a_game_summary_until_minute)

    game_summaries.append(a_game_summaries)

with open(f"{folder_path_gradThesis_data}//game_summaries.pkl", "wb") as file:  # "wb" 모드는 쓰기 및 바이너리 모드
    pickle.dump(game_summaries, file)
with open(f"{folder_path_gradThesis_data}//fail_to_load_games.pkl", "wb") as file:  # "wb" 모드는 쓰기 및 바이너리 모드
    pickle.dump(fail_to_load_games, file)

print(f"elapsed time : {time.time() - start_time}")
```
